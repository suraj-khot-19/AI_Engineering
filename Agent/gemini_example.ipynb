```python
from dotenv import load_dotenv
import os
from google import genai
from openai import OpenAI
```

```python
load_dotenv()

api_key = os.getenv("GEMINI_API_KEY")
if not api_key:
    raise RuntimeError(
        "Missing GEMINI_API_KEY. Set it in your environment (or create a .env file) before running this notebook."
    )

client = genai.Client(api_key=api_key)
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explain Spring Boot simply"
)

print(response.text)
```

```python
#task

messages = {
            "role": "user",
            "parts": [
                {
                    "text": "Give me one business idea under 20000 RS."
                }
            ]
}
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=messages
)
print("business idea: ",response.text)
```

```python
message = [
    {
        "role": "system",
        "content":"You are a bank assistant. Your role is to solve customer problem. Your bank name is SBI. Also give to the refrences in response"
    },
    {
        "role" : "user",
        "content": "How to create a account in SBI and how to create ED and FD?"
    }
]
```

```python
#using open AI
from openai import OpenAI

client = OpenAI(api_key=api_key,base_url="https://generativelanguage.googleapis.com/v1beta/openai/")

response = client.chat.completions.create(
    model = "gemini-2.5-flash",
    messages=message
)

print(response.choices[0].message.content)
```
