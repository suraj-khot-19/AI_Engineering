```python
from dotenv import load_dotenv
import os
from google import genai
from openai import OpenAI
```

```python
load_dotenv()

api_key = os.getenv("GEMINI_API_KEY")
if not api_key:
    raise RuntimeError(
        "Missing GEMINI_API_KEY. Set it in your environment (or create a .env file) before running this notebook."
    )

client = genai.Client(api_key=api_key)
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explain Spring Boot simply"
)

print(response.text)
```
