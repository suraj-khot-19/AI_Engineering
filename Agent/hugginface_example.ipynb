```python
import os 
from openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()

huggingface_api_key = os.getenv("huggingface_api_key")

if not huggingface_api_key:
    raise ValueError("huggingface_api_key is not set")
else:
    print("huggingface_api_key is set")
```

```python
openai_client = OpenAI(api_key=huggingface_api_key,base_url="https://router.huggingface.co/v1")
```

```python
messages = [
    {"role": "system", "content": "You are a skin specialist doctor. Solve users skin problems. If user query is not related to skin, say you are not able to answer that question."},
    {"role": "user", "content": "What is the capital of France?"},
    {"role": "user", "content": "Why pimple is formed? How to solve it?"}
]
```

```python
completion = openai_client.chat.completions.create(
    model="openai/gpt-oss-20b:groq",
    messages=messages
)

intruction = {
    "role":"system",
    "content":"You are a AI assitent you task to format this message as a well designed pattern. Use bullets or list and new line, bold text when nedded"
}

finalres = openai_client.chat.completions.create(
    model="openai/gpt-oss-20b:groq",
    messages=[
        intruction,
        {
            "role":"user",
            "content":completion.choices[0].message.content
        }
    ]
)

print(finalres.choices[0].message)
```
